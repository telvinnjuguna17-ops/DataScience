```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the Describe function.
data.describe()
```

```python
# We have noticed that the column item_price conatains 3 empty slots. We need to fill those empty slots with the mean of the column.
data["itemprice"] = data["itemprice"].fillna(55)

# describe the data again
data.describe()
```

```python
# get the number of enqries in a given column
data.groupby("city").size()
```

```python
# Combining multiple columns 
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a countplot of the distribution of males to females
sns.countplot(x="gender", data=data, palette="autumn")
plt.title("Gender Distribution against the number of Customers")
plt.ylabel("Total Number of Customers")
plt.xlabel("Gender")
```

```python
# Which product category appears to be the most popular?
sns.countplot(x="product", data=data, palette="autumn")
plt.title("Product Category Distribution against the number of Customers")
plt.ylabel("Total Number of Customers")
plt.xticks(rotation=60)
```

```python
# Below is a countplot that has two columns. 
sns.countplot(x="gender", hue="city", data=data, palette="autumn")
```

```python
# below is a pieplot
data["payment"].value_counts().plot.pie(autopct="%1.2f%%", colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"])
plt.title("Distribution of Payment Methods")
```
